# Introducción

El análisis de sentimiento permite interpretar automáticamente la opinión expresada por los usuarios en textos no estructurados.

En este proyecto se propone un esquema de métricas para evaluar un sistema de análisis de sentimiento aplicado al soporte al cliente en Twitter.

### _Caso de uso_

Una empresa recibe miles de consultas diarias mediante Twitter relacionadas con soporte técnico, facturación y reclamos.

El objetivo es detectar automáticamente consultas críticas y generar indicadores útiles para el área de soporte.

```python
import pandas as pd

metricas = pd.DataFrame({

    "Métrica":[
        "Accuracy",
        "Tiempo de respuesta",
        "Cobertura",
        "Índice de sentimiento",
        "Consultas críticas"
    ],

    "Objetivo":[
        "Evaluar la calidad del modelo",
        "Reducir tiempos de respuesta",
        "Analizar la mayor cantidad posible de consultas",
        "Medir la percepción de los clientes",
        "Priorizar reclamos"
    ]

})

metricas
```

# Justificación de las métricas

**Accuracy**

Permite conocer qué porcentaje de las clasificaciones realizadas por el sistema coincide con la clasificación esperada.

**Tiempo de respuesta**

Mide la rapidez con la que el sistema identifica consultas críticas.

**Cobertura**

Indica el porcentaje de consultas que pudieron analizarse automáticamente.

**Índice de sentimiento**

Permite monitorear la satisfacción general de los clientes.

**Consultas críticas**

Facilita priorizar reclamos y asignar recursos de soporte.

# Método de recolección de datos

La información proviene de las consultas realizadas por los clientes mediante Twitter.

El pipeline de procesamiento realiza automáticamente:

- Recepción de tweets.
- Limpieza del texto.
- Análisis de sentimiento.
- Clasificación automática.
- Generación de indicadores.

# Impacto en el negocio

La implementación de este sistema permite reducir tiempos de respuesta, automatizar tareas repetitivas y detectar rápidamente consultas críticas.

Además, facilita la toma de decisiones mediante indicadores que reflejan el estado del servicio y la satisfacción de los clientes.

# Escalabilidad

El pipeline puede procesar miles de consultas diariamente.

En un entorno productivo sería recomendable utilizar modelos basados en Transformers y desplegar el sistema mediante APIs o servicios en la nube para mejorar el rendimiento.

# Limitaciones

El análisis de sentimiento puede verse afectado por sarcasmos, ironías o expresiones ambiguas.

Asimismo, las métricas propuestas deben complementarse con evaluaciones periódicas para asegurar la calidad del sistema.
