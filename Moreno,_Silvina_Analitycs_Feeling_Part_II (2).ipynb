# Introducción

El análisis de sentimiento permite interpretar automáticamente la opinión expresada por los usuarios en textos no estructurados.

En este proyecto se propone un esquema de métricas para evaluar un sistema de análisis de sentimiento aplicado al soporte al cliente en Twitter.

### _Caso de uso_

Se analiza el escenario de una empresa que recibe miles de consultas diarias mediante Twitter relacionadas con soporte técnico, facturación y reclamos.

Debido al gran volumen de mensajes, resulta necesario automatizar el análisis de sentimiento para detectar consultas críticas, priorizar reclamos y generar indicadores que faciliten la gestión del área de soporte.

### _Objetivos_

- Detectar automáticamente consultas con sentimiento negativo.
- Priorizar reclamos críticos.
- Reducir el tiempo de respuesta.
- Medir la calidad del servicio.
- Automatizar el procesamiento de consultas.

```python
import pandas as pd

metricas = pd.DataFrame({
    "Métrica": [
        "Accuracy",
        "Precision",
        "Recall",
        "F1-Score",
        "Tiempo de respuesta",
        "Cobertura",
        "Índice de sentimiento"
    ],
    "Objetivo": [
        "Evaluar la exactitud del modelo",
        "Medir la calidad de las predicciones positivas",
        "Detectar correctamente consultas críticas",
        "Equilibrar Precision y Recall",
        "Reducir demoras",
        "Analizar la mayor cantidad de consultas",
        "Medir la satisfacción general"
    ]
})

metricas
```

# Fórmulas de evaluación

## Accuracy

Accuracy = (TP + TN) / (TP + TN + FP + FN)

Mide el porcentaje de clasificaciones correctas realizadas por el modelo.

---

## Precision

Precision = TP / (TP + FP)

Indica qué porcentaje de las predicciones positivas fueron correctas.

---

## Recall

Recall = TP / (TP + FN)

Evalúa la capacidad del modelo para detectar correctamente los casos positivos.

---

## F1-Score

F1 = 2 × (Precision × Recall) / (Precision + Recall)

Representa el equilibrio entre Precision y Recall y es especialmente útil cuando las clases se encuentran desbalanceadas.

# Justificación de las métricas

**Accuracy**

Permite conocer qué porcentaje de las clasificaciones realizadas por el sistema coincide con la clasificación esperada.

**Tiempo de respuesta**

Mide la rapidez con la que el sistema identifica consultas críticas.

**Cobertura**

Indica el porcentaje de consultas que pudieron analizarse automáticamente.

**Índice de sentimiento**

Permite monitorear la satisfacción general de los clientes.

**Consultas críticas**

Facilita priorizar reclamos y asignar recursos de soporte.

## _Resuemn_

Aunque Accuracy es una métrica ampliamente utilizada, puede resultar engañosa cuando existen clases desbalanceadas.

En atención al cliente suele haber muchas consultas normales y pocas consultas críticas. En este escenario, un modelo podría obtener una Accuracy elevada sin identificar correctamente los reclamos importantes.

Por este motivo resulta recomendable utilizar F1-Score, ya que combina Precision y Recall y proporciona una evaluación más equilibrada del rendimiento del modelo.

# Justificación técnica

Para este caso de uso se recomienda utilizar modelos basados en Transformers.

Las técnicas basadas en reglas y TextBlob presentan un bajo costo computacional y son sencillas de implementar, pero suelen fallar al interpretar sarcasmos, negaciones y expresiones informales, muy frecuentes en Twitter.

Los modelos Transformer, como BERT o XLM-RoBERTa, analizan el contexto completo de la oración, logrando una clasificación más precisa de las consultas.

Aunque requieren mayor capacidad computacional, ofrecen mejores resultados cuando el volumen de información y la complejidad del lenguaje aumentan.

# Consideraciones Éticas y Mitigación de Riesgos

Los sistemas de análisis de sentimiento procesan información generada por usuarios, por lo que es fundamental garantizar un uso responsable de los datos.

## Riesgos

- Interpretación incorrecta de sarcasmos o ironías.
- Sesgos del modelo hacia determinados idiomas o formas de expresión.
- Uso de información personal publicada por los usuarios.
- Clasificación errónea de consultas críticas.

## Estrategias de mitigación

- Anonimizar nombres de usuario e identificadores personales.
- Procesar únicamente la información necesaria para el análisis.
- Monitorear periódicamente el rendimiento del modelo.
- Incorporar supervisión humana en consultas críticas.
- Actualizar el modelo con nuevos datos para reducir sesgos.

Los modelos basados en Transformers ofrecen una mayor precisión gracias a su capacidad para comprender el contexto, aunque requieren una infraestructura computacional más robusta.

# Procesamiento multilingüe

Una empresa internacional puede recibir consultas en distintos idiomas.

Existen tres estrategias principales:

- utilizar un modelo específico para cada idioma;
- traducir automáticamente todos los textos antes del análisis;
- emplear modelos multilingües como mBERT o XLM-RoBERTa.

Los modelos multilingües permiten trabajar con un único pipeline, aunque requieren una mayor capacidad de procesamiento.
